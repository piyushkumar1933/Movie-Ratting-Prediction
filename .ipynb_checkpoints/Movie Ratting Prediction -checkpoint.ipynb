### Tokenizing and stemming

```python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
tokenizer = RegexpTokenizer(r'\w+') # Select all the words
ps = PorterStemmer()
en_sw = set(stopwords.words('english'))
```

```python
def getStemmedReview(review):
    #convert all to lower case
    review = review.lower()
    #replace all the br with space
    review = review.replace('<br /><br />',' ')
    #Tokenize 
    #Remove the digit contained word as wwll
    review=re.sub(r"[A-Za-z]+\d+|\d+[A-Za-z]+",'',review).strip()
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in en_sw]
    ste_tokens = [ps.stem(token) for token in new_tokens]
    
    cleaned_review = ' '.join(ste_tokens)
    return cleaned_review
```

```python
data.review = data.review.apply(getStemmedReview)
```

```python
data.head(10)
```

```python
X = data.review.values
y = data.label.values
```

### Spliting the datasets into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
trainx,testx,trainy,testy = train_test_split(X,y,test_size=0.2,random_state=9)
print("Train shapes : X = {}, y = {}".format(trainx.shape,trainy.shape))
print("Test shapes : X = {}, y = {}".format(testx.shape,testy.shape))
```

### Building Model

```python
def train(X,Y):
    class_items = {}
    class_priors = {}
    word_counts = {}
    vocab = set()
    
    n = len(X)
    grouped_data = group_by_class(X,Y)
    
    for c,data in grouped_data.items():
        class_items[c] = len(data)
        class_priors[c] = math.log(class_items[c]/n)
        word_counts[c] = defaultdict(lambda:0)
        
        for text in data:
            counts = Counter(text.split())
            for word, count in counts.items():
                if word not in vocab:
                    vocab.add(word)
                word_counts[c][word] +=count
    return class_items,class_priors,word_counts,vocab
            
    
```

```python
def group_by_class(X,Y):
    classes = np.unique(Y)
    data = {}
    for c in classes:
         data[c] = X[np.where(Y == c)]
    return data
```

```python
def predict(X,class_items,class_priors,word_counts,vocab):
    results = []
    
    for text in X:
        class_score = {}
        for c,prior in class_priors.items():
            class_score[c] = prior
        words = text.split()
        for word in words:
            if word not in vocab:
                continue
            for c in class_score.keys():
                log_w_given_c = laplasSmoothing(word,c,class_items,word_counts,vocab)
                class_score[c] +=log_w_given_c
        results.append(max(class_score,key = class_score.get))
    return results
                
```

```python
def laplasSmoothing(word,c,class_items,word_counts,vocab):
    num = word_counts[c][word]+1
    denom = class_items[c]+len(vocab)
    return math.log(num/denom)
```

```python
# First train the model
class_items,class_priors,word_counts,vocab = train(trainx[:10000],trainy[:10000]) #Memory error tahts why i have taken just the 10000 datapoints
```
